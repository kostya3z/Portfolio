### Анализ пользователей по метрикам

#### Retention Rate 

Для расчета Retention rate необходимо дополнить наш датасет информацией о дате первого посещения пользователем приложения. Данная дата ляжет в основу разбиения пользователей на когорты в будущем.

```python
# создаем отдельный датафрейм, в котором находим дату первого посещения приложения каждым пользователем
df_first_visit = final_df.groupby('user_id').agg({'event_time':'min'})\
                         .reset_index().rename(columns={'event_time':'first_visit'})
df_first_visit['first_visit'] = df_first_visit['first_visit'].round('d')
# присоединяем новый столбец к нашему датасету
final_df = final_df.merge(df_first_visit, on = 'user_id', how = 'left')

```

```python
# вычисляем лайфтайм для каждого события в днях
final_df['lifetime'] = (
final_df['event_time'] - final_df['first_visit'])


final_df['lifetime'] = final_df['lifetime']
final_df['lifetime'] = final_df['lifetime']/np.timedelta64(1,'D')
final_df['lifetime'] = final_df['lifetime'].astype(int)
```

```python
# строим таблицу удержания
retention_table = final_df.pivot_table(
    index = 'first_visit', columns = 'lifetime', values = 'user_id', aggfunc = 'nunique'
)
retention_table.head()
```

```python
# вычисляем размеры когорт

cohort_sizes = (
    final_df.groupby('first_visit')
    .agg({'user_id': 'nunique'})
    .rename(columns={'user_id': 'cohort_size'})
)
```

```python
# объединяем размеры когорт и таблицу удержания

retention_table_merged = cohort_sizes.merge(
    retention_table, on='first_visit', how='left'
).fillna(0)

retention_table_merged.head()
```

Теперь необходимо расчитать непосредственно сам Retention rate.

```python
# делим данные таблицы удержания на размеры когорт и затем убираем колонку с размером когорт

retention_table_merged = retention_table_merged.div(
    retention_table_merged['cohort_size'], axis=0
).drop(columns=['cohort_size'])

retention_table_merged.head()
```

Визуализируем полученные данные. Для этого построим heatmap.

```python
plt.figure(figsize=(20, 10))  # задаём размер графика
ax = sns.heatmap(
    retention_table_merged.drop(columns=[0]), # убираем нулевой лайфтайм
    annot=True,  # включаем подписи
    fmt='.2%',  # переводим значения в проценты
)
ax.figure.set_size_inches(25,13)
plt.title('Тепловая карта удержания')  # название графика
plt.xlabel('lifetime')
plt.ylabel('Дата первого посещения')
ticklabels = [retention_table_merged.index[int(tick)].strftime('%Y-%m-%d') for tick in ax.get_yticks()]
ax.set_yticklabels(ticklabels);
```

Ожидаемо видим, что с длительностью лайтайма Retention rate имеет тренд на снижение по когортам. Соответственно, на выходе получились корректные данные.     
Интересные наблюдения из графика:
- более поздние когорты зрительно преимущественно имеют более низкий Retention rate, особенно на первый день лайфтайма. Их цветовая гамма менее яркая по сравнению с более ранними когортами. Можно предположить, что такая тенденция продолжится далее.

#### Время проведенное пользователями в приложении

Для расчета времени определим id для каждой сессии. За границу между сессиями возьмем 10 минут.     

Полагаю, что 10 минут для тайм-аута для разделения сессий является наиболее оптимальным вариантом, так как пользователь даже при активном использовании приложения в рамках одной сессии может быть вынужен отвлечься на короткий промежуток времени. Возвращение пользователя после более длительного перерыва, представляется, уже мало вероятно будет относиться к одной и той же сессии.

```python
# предварительно отсортируем наш датасет по user_id и event_time, чтобы сесси выделились корректно
final_df = final_df.sort_values(['user_id', 'event_time']).reset_index(drop=True)
```

```python
# определяем разницу между сессиями в 10 минут
g = (final_df.groupby('user_id')['event_time'].diff() > pd.Timedelta('10Min')).cumsum()
final_df['session_id'] = final_df.groupby(['user_id', g], sort=False).ngroup() + 1
```

Теперь мы можем посмотреть, сколько времени пользователи провели в приложении, принимаяя во внимание, что одна сессия составляем 10 минут, и мы можем рассчитывать количество сессий у одного пользователя.

```python
#группируем, получаем описательные данные и строим визуализацию распределения количества сессий по пользователям
df_user_grouped = final_df.groupby('user_id').agg({'session_id':'count'})\
                          .sort_values(by = 'session_id', ascending = False)
print(df_user_grouped['session_id'].describe())

df_user_grouped['session_id'].hist(bins=100)
plt.title('Распределение количества сессий по пользователям')
plt.xlabel('Количество сессий')
plt.ylabel('Количество пользователей')
plt.show()
```

Рассчитаем время каждой сессии.

```python
# создаем новый столбец только с временем
final_df['event_timestamp'] = 0
for i in range(len(final_df['event_time'])):
    final_df['event_timestamp'][i] = final_df['event_time'][i].strftime('%Y-%m-%d %H:%M:%S') 

    
```

```python
final_df['event_timestamp'] = pd.to_datetime(final_df['event_timestamp'])
```

```python
# создаем отдельные датафреймы, в которых рассчитываем максимальную и минимульную дату и время для каждой сессии
final_df_sessionstart = final_df.groupby('session_id')\
                               .agg({'event_timestamp':'min'})\
                               .reset_index().rename(columns={'event_timestamp':'start'})
                               
final_df_sessionend = final_df.groupby('session_id')\
                               .agg({'event_timestamp':'max'})\
                               .reset_index().rename(columns={'event_timestamp':'end'})
# объединяем полученыне датафреймы в одну таблицу
final_df_sessiontime = final_df_sessionstart.merge(final_df_sessionend, on = 'session_id')

# смотрим, что получилось
final_df_sessiontime.head()
```

```python
# находим разницу между датами
final_df_sessiontime['time'] = final_df_sessiontime['end'] - final_df_sessiontime['start']
final_df_sessiontime['time'].describe()
```

Видим, что у нас нет аномально больших значений длины сесиии, 25% процентов всех сессий составляет 0 секунд.     
Посмотрим на сессии, которые составляют 0 секунда, какие именно события происходили в них.

```python
short_sessions_id = final_df_sessiontime.query('time == 0')['session_id'].unique()
```

```python
final_df.query('session_id in @short_sessions_id')['event_name'].value_counts()
```

Видим, что большая часть всех событий на такие короткие сессии приходится на просмотр подсказок и фотографий.

В имеющемся виде мы не сможем визуализировать данные, чтобы посмотреть на них более детально. Необходимо изменить тип столбца `time`.

```python
for i in range(len(final_df_sessiontime['time'])):
    final_df_sessiontime['time'][i] = final_df_sessiontime['time'][i].total_seconds()
```

```python
plt.figure(figsize=(12,7))
sns.boxplot(final_df_sessiontime['time'])
plt.title('Распределение количества сессий по времени')
plt.xlabel('Время')
plt.ylabel('Количество сессий')
plt.show()
```

Видим, что у нас есть выбивающие значения, однако они не превышают 12000 секунд (примерно 3 часа 12 минут), то есть, в целом, это адекватная длина сессии.       
Учитывая, что в нашем датасете есть данные о миллисикундах в времени посещения приложения пользователями, и как мы выяснили выше, почти 25% составляют мгновенные сессии, полагаю, без дополнительных консультаций с техническими специалистами не следует удалять такие сессии. ## Проверка гипотез

### Проверка обязательной гипотезы

Нам необходимо проверить следующую гипотезу.       
Некоторые пользователи установили приложение по ссылке из yandex, другие — из google: две эти группы демонстрируют разную конверсию в просмотры контактов.      
Сформулируем нулевую и альтернативную гипотезы.      
H0: между пользователями, которые установили приложение по ссылке из google, и пользователями, которые пришли через ссылку из yandex, отсутствуют статистически значимые отличия.     
H1:между пользователями, которые установили приложение по ссылке из google, и пользователями, которые пришли через ссылку из yandex, есть статистически значимые отличия.       

Проверять данную гипотезу будем с помощью z-теста о равенстве долей, который используется для определения статистически значимых отличий между группами при сравнении конверсий.

Для проверки гипотезы необходимо сначала подготовить данные. Количество уникальных пользователей в первых и второй группах, и сколько из них совершили целевое действие - `contacts_show`. 
Первая группа будет группами с пользователями, которые пришли из google, вторая - из yandex.

```python
# создаем датафреймы с пользователями по группам
final_df_google = final_df.loc[final_df['source'] == 'google']
final_df_yandex = final_df.loc[final_df['source'] == 'yandex']
```

```python
# находим общее количество уникальных пользователей в 1й группе
final_df_google_total = final_df_google['user_id'].nunique()
# находим количество уникальных пользователей совершихших целевое действие в 1й группе
final_df_google_cr = final_df_google.loc[final_df_google['event_name'] == 'contacts_show']['user_id'].nunique()

print('Конверсия у 1й группы составляет -', round((final_df_google_cr / final_df_google_total) * 100, 2), '%')

# находим общее количество уникальных пользователей в 2й группе
final_df_yandex_total = final_df_yandex['user_id'].nunique()
# находим количество уникальных пользователей совершихших целевое действие в 2й группе
final_df_yandex_cr = final_df_yandex.loc[final_df_yandex['event_name'] == 'contacts_show']['user_id'].nunique()

print('Конверсия у 2й группы составляет -', round((final_df_yandex_cr / final_df_yandex_total) * 100, 2), '%')

#создаем массивы с данными для проведения z-теста
purchases = np.array([final_df_google_cr, final_df_yandex_cr])
leads = np.array([final_df_google_total, final_df_yandex_total])
```

Видим, что конверсия между этими двумя группами почти одинаковая. Весьма вероятно, что и статистический тест укажет на то, что между двумя этими группами нет статистически значимых отличий.

```python
alpha = 0.05  


p1 = purchases[0] / leads[0]
p2 = purchases[1] / leads[1]



p_combined = (purchases[0] + purchases[1]) / (leads[0] + leads[1])


difference = p1-p2


z_value = difference / mth.sqrt(p_combined * (1 - p_combined) * (1/leads[0] + 1/leads[1]))

distr = st.norm(0,1)

p_value = (1 - distr.cdf(abs(z_value))) * 2



print('p-значение: ', p_value)

if p_value < alpha:
    print('Отвергаем нулевую гипотезу: между долями есть значимая разница')
else:
    print(
        'Не получилось отвергнуть нулевую гипотезу, нет оснований считать доли разными'
    )
```

Как и ожидалось, результаты теста говорят о том, нет оснований считать, что между двумя выделенными группами есть статистически значимые отличия.     
Соответственно, пользователи, установившие приложения по ссылкам из google и yandex, одинаково доходят до выполнения целевого действия.

```python
#Создадим отдельные списки пользователей, которые совершали tips_show и tips_click 
tips_show_user_list = final_df.loc[final_df['event_name'] == 'tips_show']['user_id'].unique()
tips_click_user_list = final_df.loc[final_df['event_name'] == 'tips_click']['user_id'].unique()
```

```python
# создаем датафреймы с данными пользователей, которые совершали несколько действий или только одно
final_df_1 = final_df.query('user_id in @tips_show_user_list and user_id in @tips_click_user_list')
final_df_2 = final_df.query('user_id in @tips_show_user_list and user_id not in @tips_click_user_list')
```

```python
# находим общее количество уникальных пользователей в 1й группе
final_df_1_total = final_df_1['user_id'].nunique()
# находим количество уникальных пользователей совершихших целевое действие в 1й группе
final_df_1_cr = final_df_1.loc[final_df_1['event_name'] == 'contacts_show']['user_id'].nunique()

print('Конверсия у 1й группы составляет -', round((final_df_1_cr / final_df_1_total) * 100, 2), '%')

# находим общее количество уникальных пользователей в 2й группе
final_df_2_total = final_df_2['user_id'].nunique()
# находим количество уникальных пользователей совершихших целевое действие в 2й группе
final_df_2_cr = final_df_2.loc[final_df_2['event_name'] == 'contacts_show']['user_id'].nunique()

print('Конверсия у 2й группы составляет -', round((final_df_2_cr / final_df_2_total) * 100, 2), '%')

#создаем массивы с данными для проведения z-теста
purchases = np.array([final_df_1_cr, final_df_2_cr])
leads = np.array([final_df_1_total, final_df_2_total])
```

Видим, что конверсия у первой группы пользователей, которые выполнили оба действия (`tips_show` и `tips_click`), почти в 2 раза выше, чем у второй, которые выполнили только одно дейтвие (`tips_show`).          
Представляется, что подсказки для пользователей настроены все-таки достаточно хорошо, так как пользователи, которые ими воспользовались, 2 раза чаще доходят до выполнения целевого действия.  

```python
# провдим z-тест для первой гипотезы
alpha = 0.05  


p1 = purchases[0] / leads[0]
p2 = purchases[1] / leads[1]

p_combined = (purchases[0] + purchases[1]) / (leads[0] + leads[1])

difference = p1-p2


z_value = difference / mth.sqrt(p_combined * (1 - p_combined) * (1/leads[0] + 1/leads[1]))

distr = st.norm(0,1)

p_value = (1 - distr.cdf(abs(z_value))) * 2

print('p-значение: ', p_value)

if p_value < alpha:
    print('Отвергаем нулевую гипотезу: между долями есть значимая разница')
else:
    print(
        'Не получилось отвергнуть нулевую гипотезу, нет оснований считать доли разными'
    )
```

 Видим, что между двумя выделенными ранее группами есть статистически значимые различия конверсий в целевое действие. Соответственно, можно с большей уверенностью говорить о том, что подсказки для пользователей работают достаточно хорошо.

Теперь проверим следующую гипотезу: конверсия в добавление объявления в избранное между пользователями, которые пользовались приложением в будние дни недели, и которые пользовались в выходыне дни отличается.      
Сформулируем нулевую и альтернативную гипотезы.       
Н0: нет статистически значимых различий в конверсии у данных двух групп пользователей          
Н1: есть статистически значимые различия в в конверсии у данных двух групп пользователей

Для проверки данной гипотезы также будем использовать z-тест о равенстве долей, так как мы снова проверяем гипотезу с конверсией.     
Пользователи будних дней будут входить в первую группу, пользователи выходный дней в вторую.

Датафреймы с пользователи будних и выходных дней мы уже формировали ранее. Осталось только с них собрать необходимые для теста данные.

```python
# аналогичным образом собираем данные для теста, как и ранее

# находим общее количество уникальных пользователей в 1й группе
final_df_weekday_1_total = final_df_weekday['user_id'].nunique()
# находим количество уникальных пользователей совершихших целевое действие в 1й группе
final_df_weekday_1_cr = final_df_weekday.loc[final_df_weekday['event_name'] == 'contacts_show']['user_id'].nunique()

print('Конверсия у 1й группы составляет -', round((final_df_weekday_1_cr / final_df_weekday_1_total) * 100, 2), '%')

# находим общее количество уникальных пользователей в 2й группе
final_df_weekend_2_total = final_df_weekend['user_id'].nunique()
# находим количество уникальных пользователей совершихших целевое действие в 2й группе
final_df_weekend_2_cr = final_df_weekend.loc[final_df_weekend['event_name'] == 'contacts_show']['user_id'].nunique()

print('Конверсия у 2й группы составляет -', round((final_df_weekend_2_cr / final_df_weekend_2_total) * 100, 2), '%')

#создаем массивы с данными для проведения z-теста
purchases = np.array([final_df_weekday_1_cr, final_df_weekend_2_cr])
leads = np.array([final_df_weekday_1_total, final_df_weekend_2_total])
```

Видим, что конверсия у 2й группы меньше, несмотря на то, что ранее на графиках было заметно, что в выходные пользователи более активно используют функционал приложения и совершают больше поисковых действий. Возможно из-за этого, пользователи более строго подходят к выбору предметов для сохранения объявлений в избранное.

```python
# провдим z-тест для второй гипотезы
alpha = 0.05  


p1 = purchases[0] / leads[0]
p2 = purchases[1] / leads[1]

p_combined = (purchases[0] + purchases[1]) / (leads[0] + leads[1])

difference = p1-p2


z_value = difference / mth.sqrt(p_combined * (1 - p_combined) * (1/leads[0] + 1/leads[1]))

distr = st.norm(0,1)

p_value = (1 - distr.cdf(abs(z_value))) * 2 

print('p-значение: ', p_value)

if p_value < alpha:
    print('Отвергаем нулевую гипотезу: между долями есть значимая разница')
else:
    print(
        'Не получилось отвергнуть нулевую гипотезу, нет оснований считать доли разными'
    )
```

Результаты теста показывают, что между ранее рассмотренными показателями конверсии есть статистически значимая разница.     
Данные показывают, что конверсия группы выходных пользователей устойчиво хуже пользователей будних дней. Таким образом, следует больше работать с пользователями, которые пользуются приложением в выходные дни, например, направлять им больше пуш уведомлений с рекомендованными объявлениями и напоминаний.
